```python
import numpy as np
import pandas as pd
from sympy.physics.quantum.cg import CG
```

```python
# Function to upload SU(3) Wigner coefficients from the program's output format
# path is the direction of the file
# l0, m0  are the labels of the irrep of the one-body operator coupling

def load_SU3_Wigner_file(path,name):
    
    # Filename depending on irrep
    filename = path + name

    # Open file
    with open(filename + '.txt') as f:
        lines = f.readlines()
        
    # SU3 irreps loaded
    l1, m1 = int(lines[2].split(" ")[0]), int(lines[2].split(" ")[1]) 
    l2, m2 = int(lines[6].split(" ")[0]), int(lines[6].split(" ")[1]) 
    l12, m12 = int(lines[10].split(" ")[0]), int(lines[10].split(" ")[1]) 
    
    # Values and labels stored
    values = [list(filter(lambda x: x != '', l.split(" "))) for l in lines[15:]]
    
    # Column names and multiplicidades
    cols = ["k1", "L1", "k2", "L2", "k3", "L3"]
    rhos = []
    for i in range(1,len(values[0])-5):
        cols.append("rho{0}".format(i))
        rhos.append("rho{0}".format(i))
        
    # Values stored
    values_df = pd.DataFrame(np.array(values), columns = cols)
    values_df[["k1", "L1", "k2", "L2", "k3", "L3"]] = values_df[["k1", "L1", "k2", "L2", "k3", "L3"]].astype(int)
    values_df[rhos] = values_df[rhos].astype(float)

    return SU3_Wigner(values_df)
    
# Class to store the SU(3) Wigner values 
class SU3_Wigner:
    
    def __init__(self, values):

        self.values = values
```

```python
#Load and store of files for proton excited rme
Wc01 = load_SU3_Wigner_file("./Multishell_cfp/", "coeffs_Kp=0_Lp=1")   # K=1 in Dytrych program
Wc22 = load_SU3_Wigner_file("./Multishell_cfp/", "coeffs_Kp=2_Lp=2")   # K=1 in Dytrych program
Wc03 = load_SU3_Wigner_file("./Multishell_cfp/", "coeffs_Kp=0_Lp=3")   # K=1 in Dytrych program

Wc22.values
```

```python
# Function to compute the numerator of the multi-shell cfp expansion
# lp, mlp, kp, Lp, MLp are free to choose quantum numbers
def numerator(coeffs0, lp, mlp, kp, Lp, MLp):
    
    num = 0
    coeffs = coeffs0[(coeffs0["L2"]==lp) & (coeffs0["k3"]==kp) & (coeffs0["L3"]==Lp)].reset_index()
            
    for i in range(0, len(coeffs)):
        
        # Value of SU(3) coefficient
        coeff = coeffs["rho1"][i]
        
        # Value of Clebsch-Gordan coefficient
        Lminus = coeffs["L1"][i]
        cg = float(CG(j1=Lminus, m1=MLp-mlp, j2=lp, m2=mlp, j3=Lp, m3=MLp).doit())
                
        # Sum over them
        num += coeff*cg
        
    return num

# Function to compute the multi-shell cfp 
def ms_cfp(coeffs0, lp, mlp, kp, Lp, MLp, kpp, Lpp, MLpp):
    
    coeff = coeffs0[(coeffs0["k1"]==kpp) & (coeffs0["L1"]==Lpp) & (coeffs0["k2"]==1) & (coeffs0["L2"]==lp) & (coeffs0["k3"]==kp) & (coeffs0["L3"]==Lp)]["rho1"].values[0]
    cg = float(CG(j1=Lpp, m1=MLpp, j2=lp, m2=mlp, j3=Lp, m3=MLp).doit())
    num = numerator(coeffs0, lp, mlp, kp, Lp, MLp)
    den = np.sqrt(2)*coeff*cg
    
    return num/den
```
